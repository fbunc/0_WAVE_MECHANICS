## Schr√∂dinger Equation introduction - Non relativistic

$$
i \hbar \frac{\partial \Psi(x,t)}{\partial t} = \left( -\frac{\hbar^2}{2m} \frac{\partial^2}{\partial x^2} + V(x,t) \right) \Psi(x,t)
$$


Let's clean the notation, to keep the intuition of this basic relation:

$$
\Psi(x,t) = \Psi
$$

$$
\frac{\partial \Psi(x,t)}{\partial t} = \Psi'
$$



$$
\frac{\partial^2 \Psi(x,t)}{\partial x^2} = \nabla^2 \Psi
$$

$$
V(x,t) \Psi = V \Psi
$$



$$
\Psi \in \mathbb{C}, \quad V \in \mathbb{R}
$$

## So the clean notation version is

$$
i \hbar \Psi' = -\frac{\hbar^2}{2m} \nabla^2 \Psi + V \Psi
$$



$$
i \hbar \Psi' = (-\frac{\hbar^2}{2m} \nabla^2  + V ) \Psi = \hat{H} \Psi
$$

 $$\hat{H} = -\frac{\hbar^2}{2m} \nabla^2  + V$$




$$
i \hbar \Psi' = -\frac{\hbar^2}{2m} \nabla^2 \Psi + V \Psi
